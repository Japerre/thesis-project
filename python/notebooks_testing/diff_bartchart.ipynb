```python
import matplotlib.pyplot as plt

def centered_bar_chart(values, labels):
    plt.bar(labels, values, color=['green' if v >= 0 else 'red' for v in values])
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add a horizontal line at y=0
    plt.ylabel('Values')
    plt.title('Centered Bar Chart')
    plt.show()

# Example usage
values = [5, -8, 12, -3, 7]
labels = ['A', 'B', 'C', 'D', 'E']
centered_bar_chart(values, labels)
```
